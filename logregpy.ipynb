```python

# make columns like wins, games played, win percentage, and before game win percentage
season_totals = df.groupby(['Team', 'Season'])[['Point differential', 'Yard differential', 'Turnover differential']].cumsum()

df[['Season Point Diff Total', 'Season Yard Diff Total', 'Season Turnover Total']] = season_totals
df['Season Wins'] = df.groupby(['Team', 'Season'])['Win?'].cumsum()
df['Games Played'] = df.groupby(['Team', 'Season'])['Week'].cumcount() + 1
df['Winning Percentage'] = df['Season Wins'] / df['Games Played']
df['BG Win Pct'] = df.groupby(['Team', 'Season'])['Winning Percentage'].shift(1)

df.tail(3)


```

```python
#Getting the previous season win pct, go to last week of each year, locate the value and map it previous season win pct for the entire year
week_17_2019_win_pct = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2020, 'Prev_Season Win Percentage'] = df[df['Season'] == 2020]['Team'].map(week_17_2019_win_pct)

week_17_2020_win_pct = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2021, 'Prev_Season Win Percentage'] = df[df['Season'] == 2021]['Team'].map(week_17_2020_win_pct)


week_18_2021_win_pct = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2022, 'Prev_Season Win Percentage'] = df[df['Season'] == 2022]['Team'].map(week_18_2021_win_pct)


week_18_2022_win_pct = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2023, 'Prev_Season Win Percentage'] = df[df['Season'] == 2023]['Team'].map(week_18_2022_win_pct)
```

```python
#same as above but with point differential
week_17_2019_point_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2020, 'Prev_Season Point Diff Total'] = df[df['Season'] == 2020]['Team'].map(week_17_2019_point_diff)

week_17_2020_point_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2021, 'Prev_Season Point Diff Total'] = df[df['Season'] == 2021]['Team'].map(week_17_2020_point_diff)


week_18_2021_point_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2022, 'Prev_Season Point Diff Total'] = df[df['Season'] == 2022]['Team'].map(week_18_2021_point_diff)


week_18_2022_point_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2023, 'Prev_Season Point Diff Total'] = df[df['Season'] == 2023]['Team'].map(week_18_2022_point_diff)

```

```python
#same as above with yard differential
week_17_2019_yard_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2020, 'Prev_Season Yard Diff Total'] = df[df['Season'] == 2020]['Team'].map(week_17_2019_yard_diff)


week_17_2020_yard_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2021, 'Prev_Season Yard Diff Total'] = df[df['Season'] == 2021]['Team'].map(week_17_2020_yard_diff)


week_18_2021_yard_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2022, 'Prev_Season Yard Diff Total'] = df[df['Season'] == 2022]['Team'].map(week_18_2021_yard_diff)


week_18_2022_yard_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2023, 'Prev_Season Yard Diff Total'] = df[df['Season'] == 2023]['Team'].map(week_18_2022_yard_diff)

```

```python
week_17_2019_turnover_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2020, 'Prev_Season Turnover Total'] = df[df['Season'] == 2020]['Team'].map(week_17_2019_turnover_diff)

week_17_2020_turnover_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2021, 'Prev_Season Turnover Total'] = df[df['Season'] == 2021]['Team'].map(week_17_2020_turnover_diff)

week_18_2021_turnover_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2022, 'Prev_Season Turnover Total'] = df[df['Season'] == 2022]['Team'].map(week_18_2021_turnover_diff)

week_18_2022_turnover_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2023, 'Prev_Season Turnover Total'] = df[df['Season'] == 2023]['Team'].map(week_18_2022_turnover_diff)
```

```python


# initializing a dictionary that will store the cumulative data 
cumulative_opponent_point_diff = {}
cumulative_opponent_yard_diff = {}
cumulative_opponent_turnover_diff = {}
# variable that keeps track of the current season
current_season = None
# iterates over each row in the season
for index, row in df.iterrows():
    #what part of the row we are looking at
    opponent_team = row['Team Played']
    season = row['Season']

    # Reset cumulative opponent statistics if the season changes
    if season != current_season:
        cumulative_opponent_point_diff = {}
        cumulative_opponent_yard_diff = {}
        cumulative_opponent_turnover_diff = {}        
        current_season = season

    # cumulates the opponent statistics week to week
    cumulative_opponent_point_diff[opponent_team] = cumulative_opponent_point_diff.get(opponent_team, 0) - row['Point differential']
    cumulative_opponent_yard_diff[opponent_team] = cumulative_opponent_yard_diff.get(opponent_team, 0) - row['Yard differential']
    cumulative_opponent_turnover_diff[opponent_team] = cumulative_opponent_turnover_diff.get(opponent_team, 0) - row['Turnover differential']
    

    # Add and rename columns of the data frame with these new statistics
    df.at[index, 'BG Opponent Season Point Diff Total'] = cumulative_opponent_point_diff[opponent_team]
    df.at[index, 'BG Opponent Season Yard Diff Total'] = cumulative_opponent_yard_diff[opponent_team]
    df.at[index, 'BG Opponent Season Turnover Total'] = cumulative_opponent_turnover_diff[opponent_team]
    
# shift by one so the model has the before the game happened info to make an accurate guess
shift_columns = ['BG Opponent Season Point Diff Total', 'BG Opponent Season Yard Diff Total', 'BG Opponent Season Turnover Total']
df[shift_columns] = df.groupby('Team Played')[shift_columns].shift(1)
```

```python
def calculate_opponent_bg_win_pct(row):
    # Find the opponent's BG Win Pct for the corresponding week
    opponent_team = row['Team Played']
    week = row['Week']
    season = row['Season']
    opponent_bg_win_pct = df.loc[(df['Team'] == opponent_team) & (df['Week'] == week) & (df['Season'] == season), 'BG Win Pct'].iloc[0]
    return opponent_bg_win_pct

# Apply the function to calculate Opponent BG Win Pct for each row
df['BG Opponent Winning Percentage'] = df.apply(calculate_opponent_bg_win_pct, axis=1)

# Function to reset Opponent BG Win Pct after every season
def reset_opponent_bg_win_pct(row):
    if row['Week'] == 1:
        # Reset Opponent BG Win Pct for the first week of each season
        return None
    else:
        return row['BG Opponent Winning Percentage']

# Reset Opponent BG Win Pct after every season
df['BG Opponent Winning Percentage'] = df.apply(reset_opponent_bg_win_pct, axis=1)

df[df['Team Played']=='Pittsburgh Steelers']
```

```python
#Getting the previous season win pct, same strategy already described, but now with opposing statistics
week_17_2019_opp_win_pct = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2020, 'Prev_Opp Season Win Percentage'] = df[df['Season'] == 2020]['Team Played'].map(week_17_2019_opp_win_pct)

week_17_2020_opp_win_pct = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2021, 'Prev_Opp Season Win Percentage'] = df[df['Season'] == 2021]['Team Played'].map(week_17_2020_opp_win_pct)

week_18_2021_opp_win_pct = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2022, 'Prev_Opp Season Win Percentage'] = df[df['Season'] == 2022]['Team Played'].map(week_18_2021_opp_win_pct)

week_18_2022_opp_win_pct = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Winning Percentage']
df.loc[df['Season'] == 2023, 'Prev_Opp Season Win Percentage'] = df[df['Season'] == 2023]['Team Played'].map(week_18_2022_opp_win_pct)

#Previous season point differential
week_17_2019_opp_point_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2020, 'Prev_Opp Season Point Diff Total'] = df[df['Season'] == 2020]['Team Played'].map(week_17_2019_opp_point_diff)

week_17_2020_opp_point_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2021, 'Prev_Opp Season Point Diff Total'] = df[df['Season'] == 2021]['Team Played'].map(week_17_2020_opp_point_diff)


week_18_2021_opp_point_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2022, 'Prev_Opp Season Point Diff Total'] = df[df['Season'] == 2022]['Team Played'].map(week_18_2021_opp_point_diff)


week_18_2022_opp_point_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Point Diff Total']
df.loc[df['Season'] == 2023, 'Prev_Opp Season Point Diff Total'] = df[df['Season'] == 2023]['Team Played'].map(week_18_2022_opp_point_diff)


#Previous season yard differential
week_17_2019_opp_yard_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2020, 'Prev_Opp Season Yard Diff Total'] = df[df['Season'] == 2020]['Team Played'].map(week_17_2019_opp_yard_diff)

week_17_2020_opp_yard_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2021, 'Prev_Opp Season Yard Diff Total'] = df[df['Season'] == 2021]['Team Played'].map(week_17_2020_opp_yard_diff)


week_18_2021_opp_yard_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2022, 'Prev_Opp Season Yard Diff Total'] = df[df['Season'] == 2022]['Team Played'].map(week_18_2021_opp_yard_diff)


week_18_2022_opp_yard_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Yard Diff Total']
df.loc[df['Season'] == 2023, 'Prev_Opp Season Yard Diff Total'] = df[df['Season'] == 2023]['Team Played'].map(week_18_2022_opp_yard_diff)


#Previous turnover total
week_17_2019_opp_turnover_diff = df[(df['Season'] == 2019) & (df['Week'] == 17)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2020, 'Prev_Opp Season Turnover Total'] = df[df['Season'] == 2020]['Team Played'].map(week_17_2019_opp_turnover_diff)

week_17_2020_opp_turnover_diff = df[(df['Season'] == 2020) & (df['Week'] == 17)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2021, 'Prev_Opp Season Turnover Total'] = df[df['Season'] == 2021]['Team Played'].map(week_17_2020_opp_turnover_diff)


week_18_2021_opp_turnover_diff = df[(df['Season'] == 2021) & (df['Week'] == 18)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2022, 'Prev_Opp Season Turnover Total'] = df[df['Season'] == 2022]['Team Played'].map(week_18_2021_opp_turnover_diff)


week_18_2022_opp_turnover_diff = df[(df['Season'] == 2022) & (df['Week'] == 18)].set_index('Team')['Season Turnover Total']
df.loc[df['Season'] == 2023, 'Prev_Opp Season Turnover Total'] = df[df['Season'] == 2023]['Team Played'].map(week_18_2022_opp_turnover_diff)

```

```python
#what first model will train and test on 
df_W1 = df[df['Week']==1]
```

```python
#what second model will train and test on
df_W2W5 = df[(df['Week'] > 1) & (df['Week'] < 6)]
```

```python
#what third model will train and test on
df_W6W18 = df[df['Week'] >= 6]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

w1_features = ['Prev_Season Win Percentage', 'Prev_Season Point Diff Total', 'Home?',
             'Prev_Season Yard Diff Total', 'Prev_Season Turnover Total', 'Prev_Opp Season Win Percentage', 'Prev_Opp Season Point Diff Total',
       'Prev_Opp Season Yard Diff Total', 'Prev_Opp Season Turnover Total']
#WEEK 2-5
features_w2w5 = ['Prev_Season Win Percentage', 'BG Win Pct','Prev_Season Point Diff Total',
             'Prev_Season Yard Diff Total', 'Prev_Season Turnover Total',  'Prev_Opp Season Win Percentage', 'Prev_Opp Season Point Diff Total', 
       'Prev_Opp Season Yard Diff Total', 'Prev_Opp Season Turnover Total','BG Opponent Winning Percentage']
#WEEK 6-18
features2 = [ 'Season Point Diff Total', 'Season Yard Diff Total','Season Turnover Total', 'BG Win Pct', 'Prev_Season Win Percentage', 'BG Opponent Season Point Diff Total',
       'BG Opponent Season Yard Diff Total','BG Opponent Season Turnover Total', 'BG Opponent Winning Percentage','Prev_Opp Season Win Percentage']

target = 'Win?'



x1_train, x1_test, y1_train, y1_test = train_test_split(df_W1[w1_features], df_W1[target], test_size=0.25, random_state=8)
x2_train, x2_test, y2_train, y2_test = train_test_split(df_W2W5[features_w2w5], df_W2W5[target], test_size = 0.25, random_state = 2)
x3_train, x3_test, y3_train, y3_test = train_test_split(df_W6W18[features2], df_W6W18[target], test_size = 0.20, random_state=5)


```

```python
model1 = LogisticRegression(random_state=8, max_iter=10000)
model2 = LogisticRegression(random_state=2,max_iter = 10000)
model3 = LogisticRegression(random_state=12,max_iter = 10000)
# Train the model on the training data
model1.fit(x1_train, y1_train)
model2.fit(x2_train, y2_train)
model3.fit(x3_train, y3_train)
# Make predictions on the test data
y1_pred = model1.predict(x1_test)
y2_pred = model2.predict(x2_test)
y3_pred = model3.predict(x3_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
accuracy1 = accuracy_score(y1_test, y1_pred)
print("Accuracy:", accuracy1)
print("Confusion Matrix:")
print(confusion_matrix(y1_test, y1_pred))
len(y1_pred)
```

```python
accuracy2 = accuracy_score(y2_test, y2_pred)
print(accuracy2)
print("Confusion Matrix:")
print(confusion_matrix(y2_test, y2_pred))
len(y2_pred)
```

```python
accuracy3 = accuracy_score(y3_test, y3_pred)
print(accuracy3)
print("Confusion Matrix")
print(confusion_matrix(y3_test, y3_pred))
len(y3_pred)
```
